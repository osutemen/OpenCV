```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the image and check this type

img = cv2.imread('images/face.jpg')
type(img)
```

```python
# Show this image
plt.imshow(img)
```

**MATPLOTLIB  --> RGB**

**OPENCV --> BGR**

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
# Resize the image

img_size = cv2.resize(fix_img, (800,600))
plt.imshow(img_size)
```

```python
img_size.shape
```

```python
# Resize according to a  constant

w_ratio = 0.5
h_ratio = 0.5

img_size_const = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
plt.imshow(img_size_const)
```

```python
#flip the image 
img_flip = cv2.flip(fix_img,-1)
plt.imshow(img_flip)
```
