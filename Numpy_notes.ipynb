This notebook contains exercises associated with images of the numpy library. 

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline 
```

```python
# Load the image
pic = Image.open('images/cat.jpg')
```

```python
# Check the type of image 
type(pic)
```

```python
# Convert to array
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
# Check the shape of array
pic_arr.shape
```

```python
#Show the image with numpy 
plt.imshow(pic_arr)
```

 When we use `pic_arr.shape`, the output is (550, 323, 3). This means the image is of size 550 by 323 pixels and has 3 color channels.

```python
# Red channel values 0-255
pic_red = pic_arr.copy()
plt.imshow(pic_red[:,:,0], cmap= 'gray')
```

```python
# Red channel values 0-255
pic_green = pic_arr.copy()
plt.imshow(pic_red[:,:,1], cmap= 'gray')
```

```python
# Blue channel values 0-255
pic_blue = pic_arr.copy()
plt.imshow(pic_blue[:,:,0], cmap= 'gray')
```

Now, We will zero out the green channel in the next image, leaving only the blue and red channels. This can also be done for the other channels.

```python
# Without green channel
pic_purple = pic_arr.copy()
pic_purple[:,:,1] =0
plt.imshow(pic_purple)
```

```python
# Only red channel
pic_red_only = pic_arr.copy()
pic_red_only[:,:,1] =0
pic_red_only[:,:,2] =0
plt.imshow(pic_red_only)
```
